```python
import math
from sympy import symbols, Eq, solve
import pandas as pd
from prettytable import PrettyTable
import matplotlib.pyplot as plt
import random

resultsgoalx = []
resultsgoaly = []
resultsmissx = []
resultsmissy = []
pointsout = []
results = []
score = []
miss = []
percentscore = []
percentsave = []

shots = range(1,101)
distances = range(3,153,3)

for d in distances:
  for i in shots:
    #d=60

  #Distance low
    dist1 = math.sqrt((d**2) + (2**2))
    dist2 = math.sqrt((dist1**2) + (3**2))
    dist3 = math.sqrt((dist1**2) + (9**2))

  #Distance high
    dist4 = math.sqrt((d**2) + (6**2))
    dist5 = math.sqrt((dist4**2) + (3**2))
    dist6 = math.sqrt((dist4**2) + (9**2))

  #Time
  #T=D/s
    speed = random.uniform(30, 36)

    timeq1 = (dist2*0.3048)/(speed)
    timeq2 = (dist3*0.3048)/(speed)
    timeq3 = (dist5*0.3048)/(speed)
    timeq4 = (dist6*0.3048)/(speed)

    pointsout.append(random.uniform(0,6))
    pointsout.append(random.uniform(18,24))


  #if statement
    if timeq1 < random.uniform(0.6,0.7):
      resultsgoalx.append(random.uniform(6,18))
      resultsgoaly.append(random.uniform(0,4))
      score.append(1)
    else:
      resultsmissx.append(random.uniform(6,18))
      resultsmissy.append(random.uniform(0,4))
      miss.append(1)

    if timeq2 < random.uniform(0.6,0.7):
      resultsgoalx.append(random.choice(pointsout))
      resultsgoaly.append(random.uniform(0,4))
      score.append(1)
    else:
      resultsmissx.append(random.choice(pointsout))
      resultsmissy.append(random.uniform(0,4))
      miss.append(1)

    if timeq3 < random.uniform(0.6,0.7):
      resultsgoalx.append(random.uniform(6,18))
      resultsgoaly.append(random.uniform(4,8))
      score.append(1)
    else:
      resultsmissx.append(random.uniform(6,18))
      resultsmissy.append(random.uniform(4,8))
      miss.append(1)

    if timeq4 < random.uniform(0.6,0.7):
      resultsgoalx.append(random.choice(pointsout))
      resultsgoaly.append(random.uniform(4,8))
      score.append(1)
    else:
      resultsmissx.append(random.choice(pointsout))
      resultsmissy.append(random.uniform(4,8))
      miss.append(1)


  perscore = (len(score)/(len(score) + len(miss)))*100
  percentscore.append(perscore)
  persave = (len(miss)/(len(score) + len(miss)))*100
  percentsave.append(persave)

  results.append({
    "distance": d,
    #"Over the bar": perover,
    #"Wide": perwide,
    "Goal": perscore,
    "Save": persave
})
#print(results)

table = PrettyTable()

# Add column headers
table.field_names = ["Distance to Goal(feet)", "% Scored", "% Saved"]

# Add rows
for result in results:
    distance = result["distance"]
    scored = result["Goal"]
    saved = result["Save"]
    #wide = result["Wide"]
    #over = result["Over the bar"]
  # Append the row data (with 5 columns)
    table.add_row([distance, scored, saved])

print(table)

plt.plot(distances, percentscore, color='green', label = '% Scored', marker = 'o')
plt.plot(distances, percentsave, color='orange', label = '% Saved', marker = 'o')
plt.legend()

plt.title("Reaction Based Simulation")
plt.xlabel('Distance to Goal (Feet)')
plt.ylabel('Outcome Percentage')
plt.grid(True)
plt.savefig("graphsimreact.png")
plt.show()
```
